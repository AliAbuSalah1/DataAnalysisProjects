####  Calling the libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.plotting.register_matplotlib_converters()
import matplotlib
print("Setup Complete")
```

```python
vgsales_ph ="vgsales.csv"
vgsales_data = pd.read_csv(vgsales_ph, index_col="Rank")
```

```python
vgsales_data.head()
```

```python
vgsales_data.info()
```

```python
vgsales_data.describe()
```

```python
vgsales_data = vgsales_data.dropna(subset=['Year'])
```

```python
vgsales_data.info()
```

```python
vgsales_data['Publisher'] = vgsales_data['Publisher'].fillna('Unknown')
```

```python
vgsales_data.info()
```

```python
top_games = vgsales_data.sort_values('Global_Sales', ascending=False).head(10)
```

```python
print(top_games [['Name', 'Platform', 'Year', 'Global_Sales']])
```

```python
publisher_seles = vgsales_data.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
publisher_seles.plot(kind='bar', title='Top 10 Publishers by Global Sales', figsize=(10, 5))
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.xlabel('Publisher')
print("الناشر الاكثر مبيعا عالميا هو Nintendo ")
plt.show()
```

```python
genre_sales = vgsales_data.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
genre_sales.plot(kind='bar', title='Sales by Genre', figsize=(10, 5), color = "teal")
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.xlabel('Genre')
plt.tight_layout()
print("🔹 ألعاب Actionهي الأعلى مبيعًا.")
plt.show()
```

```python
platform_sales = vgsales_data.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)
platform_sales.plot(kind='bar', title='Total Platform by Global Sales', figsize=(10, 5), color = "green")
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.xlabel('Platform')
print("🔹 منصة PS2 حققت أعلى مبيعات.")
plt.show()
```

```python
year_sales = vgsales_data.groupby('Year')['Global_Sales'].sum()
year_sales.plot(kind='bar', title='Global Sales Over the Years', figsize=(12, 6))
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.show()
```

```python
year_sales = vgsales_data.groupby('Year')['Global_Sales'].sum()
year_sales.plot(kind='line', marker='o', figsize=(12,6), title='Global Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
