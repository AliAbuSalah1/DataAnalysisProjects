```python
print(Students_data['Stream'].value_counts())
print(Students_data['Impact_on_Grades'].value_counts(normalize=True))
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Device_Used', data=Students_data, order=Students_data['Device_Used'].value_counts().index, palette='Set2')
plt.title('Number of students by device type')
plt.xlabel('Device used')
plt.ylabel('Number of students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#    
plt.figure(figsize=(10,5))
sns.countplot(x='Stream',data=Students_data,order=Students_data['Stream'].value_counts().index)
plt.title('Distribution of students by major')
plt.xlabel('specialty')
plt.ylabel('Number of students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#     
plt.figure(figsize=(8,5))
sns.histplot(Students_data['Daily_Usage_Hours'],kde=True , color="skyblue",bins=30)
plt.title('Distribution of the number of hours using artificial intelligence tools daily')
plt.xlabel('Number of hours')
plt.ylabel('Number of students')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x='Stream',y='Daily_Usage_Hours',data=Students_data)
plt.title('Comparison of the number of daily usage hours by specialization')
plt.xlabel('specialization')
plt.ylabel('Number of hours')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.stripplot(x='Trust_in_AI_Tools', y='Impact_on_Grades', data=Students_data, hue='Stream', jitter=0.25, dodge=True, alpha=0.7)
plt.title('The relationship between trust in tools and their impact on performance')
plt.xlabel('Trust in AI tools')
plt.ylabel('Its impact on the grades')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Trust_in_AI_Tools', y='Impact_on_Grades', data=Students_data, palette='Set3')
plt.title('Boxplot: Trust in AI Tools vs Impact on Grades')
plt.xlabel('Trust in AI tools')
plt.ylabel('Impact on Grades')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
numerical_cols = ['Daily_Usage_Hours', 'Trust_in_AI_Tools', 'Impact_on_Grades', 'Awareness_Level']
correlation = Students_data[numerical_cols].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation matrix between numerical variables')
plt.tight_layout()
plt.show()
```
