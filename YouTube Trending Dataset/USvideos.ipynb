```python
import pandas as pd
```

```python
USvideos = pd.read_csv('CAvideos.csv')
USvideos.head(10)
```

```python
USvideos.info()
```

```python
USvideos.describe()
```

```python
USvideos['category_name'] = USvideos['category_id'].map(categories)

print(USvideos[['video_id', 'title', 'channel_title', 'category_name', 'views', 'likes', 'dislikes']].head())
```

```python
print('View Categories')
top_categories = USvideos.groupby('category_name')['views'].sum().sort_values(ascending=False)
print(top_categories)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_categories = USvideos.groupby('category_name')['views'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_categories.values , y = top_categories.index , palette='viridis')
plt.title('Top 10 Categories by Total Views')
plt.xlabel('Total Views')
plt.xlabel('Category')
print("Entertainment was the most watched.")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=USvideos.sample(1000), x='views', y='likes', alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.title('Relationship between Views and Likes (Sample of 1000 videos)')
plt.xlabel('Views (log scale)')
plt.ylabel('Likes (log scale)')
plt.show()
```

```python
USvideos['publish_hour'] = pd.to_datetime(USvideos['publish_time']).dt.hour

plt.figure(figsize=(10,5))
sns.countplot(x='publish_hour', data=USvideos, palette='coolwarm')
plt.title('Distribution of Video Publish Hours')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Videos Published')
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.boxplot(x='category_name', y='views', data=USvideos)
plt.yscale('log')
plt.title('Views Distribution by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=USvideos, x='views', y='likes')
plt.title('The relationship between views and likes')
plt.xlabel('Number of views')
plt.ylabel('Number of likes')
plt.show()
```

```python
sns.countplot(data=USvideos, x='comments_disabled')
plt.title('Number of videos with comments disabled')
plt.xlabel('Are comments disabled?')
plt.ylabel('Number of videos')
plt.xticks([0,1], ['No', 'Yes'])
plt.show()
```

```python
sns.histplot(USvideos['views'], bins=100, kde=True)
plt.title('Distribution of Views')
plt.xlabel('Number of Views')
plt.ylabel('Number of Videos')
plt.xlim(0, 1_000_000) # To reduce the impact of outliers
plt.show()
```

```python
top_channels = USvideos['channel_title'].value_counts().head(10)
top_channels.plot(kind='bar', figsize=(10,5))
plt.title('Most trending channels')
plt.ylabel('Number of trending videos')
plt.xlabel('Channel name')
plt.xticks(rotation=45)
plt.show()
```

```python
USvideos['trending_date'] = pd.to_datetime(USvideos['trending_date'], format='%y.%d.%m')

daily_trend = (
    USvideos.groupby('trending_date')['video_id']
      .nunique()
      .reset_index(name='n_trending_videos')
      .sort_values('trending_date')
)

plt.figure(figsize=(14,5))
plt.plot(daily_trend['trending_date'], daily_trend['n_trending_videos'])
plt.title('Number of videos that trended daily')
plt.xlabel('Date')
plt.ylabel('Number of videos')
plt.tight_layout()
plt.show()
```

```python
USvideos['publish_dt']  = pd.to_datetime(USvideos['publish_time'])
USvideos['publish_hour'] = USvideos['publish_dt'].dt.hour
USvideos['publish_dow']  = USvideos['publish_dt'].dt.day_name()  # Monday â€¦ Sunday

heat = (
    USvideos.pivot_table(index='publish_dow', columns='publish_hour',
                   values='video_id', aggfunc='count')
      .reindex(index=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
)

plt.figure(figsize=(14,6))
sns.heatmap(heat, cmap='YlGnBu')
plt.title('Amount of trending videos by day of the week and time of publication')
plt.xlabel('Time of publication')
plt.ylabel('Day of the week')
plt.show()
```

```python
numeric_cols = ['views', 'likes', 'dislikes', 'comment_count']
corr = USvideos[numeric_cols].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation matrix between interaction indicators')
plt.show()
```

```python
USvideos['like_ratio'] = USvideos['likes'] / USvideos['views']   # Ø£Ùˆ likes / (likes+dislikes)

like_ratio_cat = (
    USvideos.groupby('category_name')['like_ratio']
      .mean()
      .dropna()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,5))
sns.barplot(x=like_ratio_cat.values, y=like_ratio_cat.index)
plt.title('Top 10 Categories by Average Like Ratio')
plt.xlabel('Average Like Ratio')
plt.ylabel('Category')
plt.show()
```
